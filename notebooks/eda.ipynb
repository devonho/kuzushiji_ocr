```python
import os
import pandas as pd
import numpy as np
from numpy import load
from matplotlib import pyplot as plt
```

```python
labels_train = load("../datasets/kuzushiji/kmnist-train-labels.npz")['arr_0']
img_train = load("../datasets/kuzushiji/kmnist-train-imgs.npz")['arr_0']
```

```python
%matplotlib inline
def show_img():
    fig = plt.Figure(figsize=[40,20], dpi=50)
    axes = fig.subplots(10,20)
    for j in range(10):
        img_train_0 = img_train[np.where(labels_train==j)]
        for i in range(20):
            axes[j,i].imshow(img_train_0[i])
    return fig

show_img()
```
